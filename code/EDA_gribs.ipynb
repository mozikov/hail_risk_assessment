## Нужно подготовить датасет для обучения: 
У нас имеются данные за каждые 6 часов;

Каждые 4 тензора соединяем в один, вот наша обучающая выборка для сетки;

```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
import glob

land_path = "data/Texas_gribs/land"
aerology_path = "data/Texas_gribs/cutted_aerology"
extra_feature_path = "data/Texas_gribs/extra_feature"
runoff_path = "data/Texas_gribs/runoff_only"
```

```python
def prepare_full_train_data(aerology_path: str,
                            land_path: str,
                            runoff_path: str,
                            #extra_feature_path: str,
                            one_day: bool = False):
    aerology_paths = glob.glob(aerology_path + "/*.grib")
    land_paths = glob.glob(land_path + "/*.grib")
    runoff_paths = glob.glob(runoff_path + "/*.grib")
    full_train_days = []
    for a_path, l_path, ro_path in zip(land_paths, aerology_paths, runoff_paths):
        full_train_days.append(np.concatenate([prepare_train_data(l_path, one_day),
                                               prepare_train_data(a_path, one_day),
                                               prepare_runoff(ro_path, one_day)], axis=1))
    #extra_feature_data = np.expand_dims(prepare_extra_feature(extra_feature_path), axis=1)
    #full_train_days.append(extra_feature_data)
    full_train_days = np.concatenate(full_train_days, axis=0)
    return full_train_days

def prepare_train_data(path: str, step: int = 6, one_day: bool = True):

    ds = xr.open_dataset(path, engine="cfgrib")
    vars = list(ds.data_vars)
    nps = {}.fromkeys(vars)

    for var in vars:
        nps[var] = ds[var].to_numpy()

    train_day = []
    time = ds.dims["time"]
    day_var = {}.fromkeys(vars)
    del ds

    for day in range(time * 6 // 24):
        for var in vars:
            day_var[var] = []
            print(var)
        for hour_period in range(4):
            for var in vars:
                day_var[var].append(nps[var][day * 4 + hour_period])
        train_day.append(np.array([day_var[var] for var in vars]).reshape(-1, 41, 65))
        if one_day:
            break

    train_day = np.array(train_day)
    return train_day

def prepare_runoff(runoff_path: str, one_day: bool = True):
    ds = xr.open_dataset(runoff_path, engine="cfgrib")
    vars = list(ds.data_vars)
    nps = {}.fromkeys(vars)

    for var in vars:
        nps[var] = ds[var].to_numpy()[:, 1, :, :]
        print(var)

    train_day = []
    time = ds.dims["time"]
    day_var = {}.fromkeys(vars)
    del ds

    for day in range(time * 12 // 24):
        for var in vars:
            day_var[var] = []
        for hour_period in range(2):
            for var in vars:
                day_var[var].append(nps[var][day * 2 + hour_period])
        train_day.append(np.array([day_var[var] for var in vars]).reshape(-1, 41, 65))
        if one_day:
            break

    train_day = np.array(train_day)
    return train_day


def prepare_extra_feature(extra_feature_path: str):
    extra_feature_paths = glob.glob(extra_feature_path + "/*.npy")
    full_ef_ds = []
    for ef_path in extra_feature_paths:
        full_ef_ds.append(np.load(ef_path))
    full_ef_ds = np.concatenate(full_ef_ds, axis=0)
    return full_ef_ds

    

train_day = prepare_full_train_data(aerology_path, land_path, runoff_path, one_day=True)
#train_day.shape
```
